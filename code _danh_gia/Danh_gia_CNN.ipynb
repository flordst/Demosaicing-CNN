TÍNH SSIM

```python
import os
import cv2
import numpy as np
from skimage.metrics import structural_similarity as ssim

def calculate_ssim(img1, img2):
    img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
    img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
    s = ssim(img1, img2)
    return s

folder_path = '/content/gdrive/MyDrive/project/deep_learning_for_demosaicing/data/data_outs'  # Thay đổi thành đường dẫn thư mục lưu kết quả của bạn trong drive
output_file = 'ssim_results.txt'      # Tên file để lưu kết quả SSIM

ssim_list = []

with open(output_file, 'w') as f:
    for i in range(407):
        original_filename = f"{i}_original.TIF"
        result_filename = f"{i}_result.TIF"

        original_path = os.path.join(folder_path, original_filename)
        result_path = os.path.join(folder_path, result_filename)

        if os.path.exists(original_path) and os.path.exists(result_path):
            img_original = cv2.imread(original_path)
            img_result = cv2.imread(result_path)

            ssim_value = calculate_ssim(img_original, img_result)
            ssim_list.append(ssim_value)

            f.write(f'Image: {i} - SSIM: {ssim_value:.4f}\n')
        else:
            f.write(f'Image: {i} - Missing original or result image\n')

ssim_average = np.mean(ssim_list)
with open(output_file, 'a') as f:
    f.write(f'Average SSIM: {ssim_average:.4f}\n')

print(f'Average SSIM: {ssim_average:.4f}')
print(f'Results saved in file: {output_file}')
```
